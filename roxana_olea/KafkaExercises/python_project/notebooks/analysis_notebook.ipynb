```python
# GRAPHIC

import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

engine = create_engine('sqlite:///market_data.db')

market_data = pd.read_sql('SELECT * FROM market_data_table', engine)

transformed_data = pd.read_sql('SELECT * FROM transformed_data_table', engine)

strategy_data = pd.read_sql('SELECT * FROM strategy_data_table', engine)

plt.figure(figsize=(14, 7))
plt.plot(market_data['timestamp'], market_data['close'], label='Close Price')
plt.title('Market Data - Close Price')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.show()

plt.figure(figsize=(14, 7))
plt.plot(strategy_data['timestamp'], strategy_data['close'], label='Close Price')
plt.plot(strategy_data['timestamp'], strategy_data['short_ma'], label='Short MA')
plt.plot(strategy_data['timestamp'], strategy_data['long_ma'], label='Long MA')
plt.scatter(strategy_data.loc[strategy_data['position'] == 1].index, 
            strategy_data.loc[strategy_data['position'] == 1]['close'], 
            label='Buy Signal', marker='^', color='g')
plt.scatter(strategy_data.loc[strategy_data['position'] == -1].index, 
            strategy_data.loc[strategy_data['position'] == -1]['close'], 
            label='Sell Signal', marker='v', color='r')
plt.title('Trading Strategy - Buy/Sell Signals')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.legend()
plt.show()
```
